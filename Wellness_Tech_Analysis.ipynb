## Users Categorization
I am going to creat a categorization of the users by Level of Physical Activity and Device Usage, following these arguments:

**Physical Activity**
* Sedentary: less than 6000 average of steps daily.
* Active: Between 6000 and 12000 average of daily steps.
* Very Active: More than 12000 average of daily steps.

**Device Usage**
* Low Use: less than 8 hours of use per day.
* Noraml Use: between 8 to 16 hours of use per day.
* High Use: more than 16 hours of use per day.

The reson of creating these categories is to know that maybe there are some users using the device for a long time daily but they are not active hoysically, also maybe there are some users active physically by using the devices for short time daily.

```python
#Categorization

#1 Physical Activity

#Grouping data by id using groupby function
id_group = df.groupby(['id'])

#Finding the steps average
id_average_steps = id_group['total_steps'].mean().reset_index()

#Categorize the physical activity
physical_bins = [0, 6000, 12000, float('inf')]
physical_labels = ['Sedentary', 'Active', 'Very Active']
id_average_steps['physical_activity'] = pd.cut(
id_average_steps['total_steps'],
bins = physical_bins,
labels = physical_labels
)

#merge physical activity into the original dataframe
df= df.merge(id_average_steps[['id', 'physical_activity']],
on='id', how='left')

#====================
#2 Device Usage

#calculate total device usage in hours per user
df['total_device_usage_hours'] = (
    df['very_active_minutes']+
    df['fairly_active_minutes']+
    df['lightly_active_minutes']+
    df['sedentary_minutes']
) / 60

#Group by user and calculate average device usage
user_device_usage = df.groupby('id')['total_device_usage_hours'].mean().reset_index()

#Categorize Device Usage
device_bins = [0, 8, 16, float('inf')]
device_labels = ['Low Use', 'Normal Use', 'High Use']
user_device_usage['device_usage'] = pd.cut(
    user_device_usage['total_device_usage_hours'],
    bins = device_bins,
    labels = device_labels
)

#Merge device usage into the original DataFrame
df = df.merge(user_device_usage[['id', 'device_usage']], on='id', how='left')


#Old categorization
'''
#Creating a data frame for the results
id_average_steps = id_average_steps.to_frame()

#check the output using the data frame
#id_average_steps

#Creating a new column that contains the average of steps for each user,
#Then I will create the categories

  #Creating list of conditions
conditions =[
    (id_average_steps <= 6000),
    (id_average_steps > 6000) & (id_average_steps < 12000),
    (id_average_steps >=12000)
]
  #Creating List of categories
categories = ['Sedentery', 'Active', 'Very Active']

#Creating a column using numpy.select() function
id_average_steps['activity_level'] = np.select(conditions, categories, default='Unknown')

#Adding the results inside a variable to make it easier to excecute
id_activity_level = id_average_steps['activity_level']

#Adding activity_level to the original dataset
df= df.merge(id_average_steps['activity_level'], on = 'id', how='left')

#id_average_steps

#2 I will do the category number 2 later on since this project is a milestone one.
#Also I want to add these new two columns in the main dataset.
'''
```

## Correlation Between Total Steps and Calories
I want to represent this correlation via scatter plot using seaborn library.

```python
# I will put this correlation inside a variable called axis,
# the scatter plot will be created using sns.scatterplot() function.
# The colouring will be by activity level using hue parameter.
#
axes = sns.scatterplot(x='total_steps', y='calories', data=df, hue='physical_activity')

#Adding the titles using plt.title() function
plt.title('Total Steps vs Calories')




plt.show()
```

### Outcomes from the previous graph

## Average Steps Per Day

```python
#Creating a list of week days
day_of_week = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday']

#Creating a figure using fig
#Adding a single set of axis using plt.subplots() function
#Sprcifying the size ininshes using figsize
fig, axes = plt.subplots(1,1,figsize=(9,6))

#Grouping data by day and calculating the averages
day_group = df.groupby(['day_of_week'])
avg_daily_steps = day_group['total_steps'].mean()

avg_steps = df['total_steps'].mean() #Overall average steps

#Iserting bar chart using plt.bar() function
plt.bar(avg_daily_steps.index, avg_daily_steps)


#Sorting the days labels to make them more readable
#Starting from Monday and ended on Sunday as I arranged them in the array
axes.set_xticks(range(len(day_of_week)))
axes.set_xticklabels(day_of_week)

# Add a horizontal line for the overall average steps
axes.axhline(y=avg_steps, color='red',label='Overall Average Steps')

#Insert Titles
axes.set_ylabel('Number of Steps')
axes.set_xlabel('Day of Week')
axes.set_title('Average Steps Per Day')

plt.legend()
plt.show()
```

### Outcomes From The Previous Graph

The users are most active on Monday, Saturday, and Tuesday. Wednesday, Thursday, and Friday are below the average but the three fell into the same area. Sunday is the least active of all the weekdays.

From this I can understand the activity of majority of users:
* I will assume that each user has 1-2 days off weekly maybe Saturday and sunday for sure, Since Sunday is the less activity day, Maybe the users like to spent their time with family and doing some homeworks.
* Users will be more active on saturdays, maybe because it's the day before Sunday (The day off for sure).


## Description of Activity Minutes

I want to explore The Percentage time of the activity for users.

```python
#Creating a Pie Chart Using matplotlib.pyplot library
#First, I want to create the categories for the graph

very_active_mins=df['very_active_minutes'].sum()
fairly_active_mins=df['fairly_active_minutes'].sum()
lightly_active_mins=df['lightly_active_minutes'].sum()
sedentary_active_mins=df['sedentary_minutes'].sum()

activity_labels = ['Very Active', 'Fairly Active', 'Lightly Active', 'Sedentary']
activity_values = [very_active_mins,
                   fairly_active_mins,
                   lightly_active_mins,
                   sedentary_active_mins
                   ]

#Creating the Pie Chart
fig, axes = plt.subplots(figsize=(8,8))
axes.pie(
    activity_values, labels=activity_labels,
    autopct = '%1.1f%%',  #display the percentage values
    startangle=140,       #Rotate the start point of the pie chart
    shadow=True,
    colors=['gold', 'lightblue', 'lightgreen', 'pink'],
    wedgeprops={'edgecolor':'black'}
)

axes.set_title('Description of Activity Minutes')

plt.legend()
plt.show
```

This pie chart shows that the users are in a sedentary state of activity most of the time, a sixth of the time doing light activity and only 2% of the time being active doing proper excercises.

* I can understand one trend from here that there is a big possible that users like to wear the bands because it's look so cool.
